```python
def MultipleFastas(filename):
    
    file = filename

    with open(file, 'r') as sequence:  #open fasta file
        data2 = sequence.readlines()

    seq = []
    full = ''

    for i in data2:                       #This iterates through the fasta (opened again to not interfere with the ids) and finds id line
        if i[0] == '>':
            unique = (i.replace(i[-1], ">"))
            seq.append(unique)           #Adds ">" at the end of the id line to make a clear separation between headers and sequences
        else:
            seq.append(i)                #Prints each line into empty newfasta list

    oneline = "".join(line.strip() for line in seq)
    id_seq = oneline.split(">")       

    nt = ['A','C','G','T']
    seqs = []

    for x in range(len(id_seq)):
        if id_seq[x].startswith('C'):
            seqs.append(id_seq[x])
        elif id_seq[x].startswith('G'):
            seqs.append(id_seq[x])
        elif id_seq[x].startswith('T'):
            seqs.append(id_seq[x])
        elif id_seq[x].startswith('A'):
            seqs.append(id_seq[x])

    return seqs

def FindSharedMotif(seqs): #seqs is list of list with first item being the motif

    import regex
    
    motif_seq = seqs[0]  #make first sequence in fasta file the motif seq and remove from fasta
    seqs.remove(seqs[0])

    substring = []

    #Incrementally append substring from each index to the end, respectively
    for i in range(len(motif_seq)):
        for j in range(len(motif_seq)):
            substring.append(motif_seq[i:j+1:])

    substring = [x for x in substring if len(x)>1]  #remove substring if less than 2 bp long

    list_substring = sorted(set(substring), key=len, reverse = False)  #order set by length and make list

    index = []
    max_len = 1

    for motif in list_substring:    #iterate through each substring increment

        if len(motif) > max_len and bool(regex.search(motif, seqs[0])) is True:       #if length of substring is greater than max_len (which is the length of the last substring that has been appended) and is in one other given sequence from fasta 
            counter = 0                                                               #this counter is initialized for each substring, keeps track of how many fasta sequences the substring is in
            for dna in seqs:                                                          #iterate through fasta sequences
                if bool(regex.search(motif, dna)) is True:                 #if substring is in each given fasta      
                    if counter < len(seqs):                                #if counter is less than the number of sequences, add 1 
                        counter = counter + 1
                        if counter == len(seqs):                           #if counter is equal to the number of sequences after adding 1, that means it is in every sequence and is a common string                    
                            max_len = len(motif)                           #reset max_len to length of substring and append to index
                            index.append(motif)

        elif len(motif) == len(index[-1])+2 and bool(regex.search(motif, seqs[0])) is False:     #if motif is 2 bp larger than the last substring appended and is not in the given sequence, that means there are no other common motifs and we can break the loop 
            break

        else:                                                              #continue lets us move to the next iteration (for speed)
            continue

    return index[-1]                                                       #return last substring appended

    #GAACCCAGGTCATCGGTTCAAGGGAAACGCTCACAATCCTCAAATAGACGCCCATGACTCACTACGTATGACAACGCACAGCGAGAAGTACGCTCGCAAAGCCTTTGGGTGTCAAATGGCGACGGGATACGGCAGTTATCCCGCT
```

```python
seqs = MultipleFastas('rosalind_lcsm.txt')   #read in file
print(FindSharedMotif(seqs))
```
