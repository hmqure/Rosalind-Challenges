```python
def GetLettersAndN(file):

    with open(file, 'r') as sequence:  #open fasta file
        data = sequence.readlines()
    
    letter_string = data[0].strip('\n')
    n = int(data[1].strip('\n')) 
    letters = letter_string.split(" ")
    
    return letters, n

def EnumerateKmers(letters, n):
    
    import random

    permlist=[]
    perm = ""
    count = 0
    count2 = 1

    permlist.append(random.choices(letters, k=n))

    while count != count2:
        for i in range(len(permlist)):
            y = random.choices(letters, k=n)
            if y not in permlist:
                permlist.append(y)
                count2 = count2 + 1
        count = count + 1

    #permlist = permlist.sort()

    final = []

    for i in permlist:
        final.append(''.join(i))

    final.sort()

    for i in final:
        print(''.join(i))

```
