```python
def GetLettersAndN(file):

    with open(file, 'r') as sequence:  #open fasta file
        data = sequence.readlines()
    
    letter_string = data[0].strip('\n')
    n = int(data[1].strip('\n')) 
    letters = letter_string.split(" ")
    
    return letters, n

def OrderLexicographically(letters, n):

    import random

    #only works in short amount of time if letters is <= 10
    #shuffling is not efficient way to get permutations 

    permlist=[]
    perm = ""

    permlist.append(random.choices(letters, k=n))

    runs = 0

    for i in range(n+1):
        runs = runs + (len(letters)**i)

    # print(len(letters))
    # print(runs)

    while len(permlist) < runs:
        for j in range(len(permlist)):
            for i in range(n+1):
                y = random.choices(letters, k=i)
                if y not in permlist:
                    permlist.append(y)

    final = []

    for i in permlist:
        final.append(''.join(i))

    for i in final:
        if i == '':
            final.remove(i)

    final = sorted(final, key=lambda word: [letters.index(c) for c in word])

    for i in final:
        print("".join(i))

#     print(len(final))
    
```

```python
output = GetLettersAndN('rosalind_lexv.txt')
OrderLexicographically(output[0], output[1])
```
