```python
def MakeListOfFasta(file):

    with open(file, 'r') as sequence:  #open fasta file
        data2 = sequence.readlines()

    seq = []
    full = ''

    for i in data2:                       #This iterates through the fasta (opened again to not interfere with the ids) and finds id line
        if i[0] == '>':
            unique = (i.replace(i[-1], ">"))
            seq.append(unique)           #Adds ">" at the end of the id line to make a clear separation between headers and sequences
        else:
            seq.append(i)                #Prints each line into empty newfasta list

    oneline = "".join(line.strip() for line in seq)
    id_seq = oneline.split(">")       

    nt = ['A','C','G','T']
    seqs = []

    for x in range(len(id_seq)):
        if id_seq[x].startswith('C'):
            seqs.append(id_seq[x])
        elif id_seq[x].startswith('G'):
            seqs.append(id_seq[x])
        elif id_seq[x].startswith('T'):
            seqs.append(id_seq[x])
        elif id_seq[x].startswith('A'):
            seqs.append(id_seq[x])

    return seqs #returns list of sequences from fasta file

def MakeConsensusProfileMatrix(seqs):
    
    length = len(seqs[0])

    fol = []

    for i in seqs:
        fol.append(list(i))

    rows, cols = 4,length

    my_matrix = [([0]*cols) for i in range(rows)]

    for j in range(len(fol)):
        for i in range(length):
            if fol[j][i] == 'A':
                my_matrix[0][i] += 1
            if fol[j][i] == 'C':
                my_matrix[1][i] += 1
            if fol[j][i] == 'G':
                my_matrix[2][i] += 1
            if fol[j][i] == 'T':
                my_matrix[3][i] += 1
   
    return my_matrix #return consensus matrix as list of lists

def MakeConsensusSequence(my_matrix):

    import numpy as np
    
    transposed = np.array(my_matrix).T.tolist()

    consensus = ''

    for i in range(len(transposed)):
        if transposed[i][0] == max(transposed[i]):
            consensus = consensus + 'A'
        elif transposed[i][1] == max(transposed[i]):
            consensus = consensus + 'C'
        elif transposed[i][2] == max(transposed[i]):
            consensus = consensus + 'G'
        elif transposed[i][3] == max(transposed[i]):
            consensus = consensus + 'T'

    return consensus

def PrintConsensusProfileMatrix(my_matrix):
    
    nuc = ['A', 'C', 'G', 'T']
    
    for i in range(len(nuc)):
        print(nuc[i]+":",*my_matrix[i])
```

```python
seqs = MakeListOfFasta('rosalind_cons.txt')
profile_matrix = MakeConsensusProfileMatrix(seqs)
consensus_string = MakeConsensusSequence(profile_matrix)

print(consensus_string)
PrintConsensusProfileMatrix(profile_matrix)
```
