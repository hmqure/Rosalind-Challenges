```python
def Getn(file):
    
    with open(file, 'r') as dna:  #open fasta file
        data = dna.readlines()
    
    n = int(data[0].strip('\n'))
    
    dna.close()  #close file
    
    return n 

def Permutate(n):

    import random

    permlist=[]
    perm = ""

    for i in range(n):
        perm = perm + str(i+1)

    permlist.append(perm)

    count = 0
    count2 = 1
    while count != count2:
        for i in range(len(permlist)):
            y = ''.join(random.sample(permlist[0],len(permlist[0])))
            if y not in permlist:
                permlist.append(y)
                count2 = count2 + 1
        count = count + 1

    print(len(permlist))
    for i in permlist:
        print(" ".join(i))
```

```python
n = Getn('rosalind_perm.txt')
Permutate(n)
```
