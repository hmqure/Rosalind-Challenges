```python
def GetSeq(file):
    
#     file = 'rosalind_orf.txt'

    with open(file, 'r') as dna:  #open fasta file
        data = dna.readlines()

    seq = ''

    for row in data[1:]:      #iterate through file after header and strip newline characters
        fasta = (row.strip('\n'))
        seq = seq + fasta     #join to empty list

    dna.close()  #close file
    
    return seq

def OpenReadingFrames(seq):
    
    aminos = {'AAA': 'K', 'AAC': 'N', 'AAG': 'K', 'AAT': 'N', 'ACA': 'T', 'ACC': 'T', 'ACG': 'T', 'ACT': 'T', 'AGA': 'R', 'AGC': 'S', 'AGG': 'R', 'AGT': 'S', 'ATA': 'I', 'ATC': 'I', 'ATG': 'M', 'ATT': 'I', 'CAA': 'Q', 'CAC': 'H', 'CAG': 'Q', 'CAT': 'H', 'CCA': 'P', 'CCC': 'P', 'CCG': 'P', 'CCT': 'P', 'CGA': 'R', 'CGC': 'R', 'CGG': 'R', 'CGT': 'R', 'CTA': 'L', 'CTC': 'L', 'CTG': 'L', 'CTT': 'L', 'GAA': 'E', 'GAC': 'D', 'GAG': 'E', 'GAT': 'D', 'GCA': 'A', 'GCC': 'A', 'GCG': 'A', 'GCT': 'A', 'GGA': 'G', 'GGC': 'G', 'GGG': 'G', 'GGT': 'G', 'GTA': 'V', 'GTC': 'V', 'GTG': 'V', 'GTT': 'V', 'TAA': '*', 'TAC': 'Y', 'TAG': '*', 'TAT': 'Y', 'TCA': 'S', 'TCC': 'S', 'TCG': 'S', 'TCT': 'S', 'TGA': '*', 'TGC': 'C', 'TGG': 'W', 'TGT': 'C', 'TTA': 'L', 'TTC': 'F', 'TTG': 'L', 'TTT': 'F'}

    orf1 = []
    orf1_proteins = ''


    for aa in range(0,len(seq),3):   #iterate through dna seq and append every three letters to new list as codons
        orf1.append(seq[aa:aa+3])

    for i in orf1:     #iterate through list of codons and store its amino acid symbol to empty string, making protein sequence
        if i in aminos.keys():
            orf1_proteins += aminos[i]

    ####### coding sequence ##########
    coding_seq = []

    for i in range(len(orf1_proteins)):
        for j in range(len(orf1_proteins)):
            if orf1_proteins[i] == 'M':
                if orf1_proteins[j] == '*':
                    coding_seq.append(orf1_proteins[i:j:])

    ######### coding sequence ##########

    orf2 = []
    orf2_proteins = ''


    for aa2 in range(1,len(seq),3):   #Repeat methodology in next ORF, but begin iterating through dna seq on the second character
        orf2.append(seq[aa2:aa2+3])

    for codon in orf2:        #remove item in list if it is not a full three letter codon
        if len(codon) != 3:
            orf2.remove(codon)

    for i2 in orf2:     #translate to proteins and store in empty string
        if i2 in aminos.keys():
            orf2_proteins += aminos[i2]

    for i in range(len(orf2_proteins)):
        for j in range(len(orf2_proteins)):
            if orf2_proteins[i] == 'M':
                if orf2_proteins[j] == '*':
                    coding_seq.append(orf2_proteins[i:j:])

    ##### ORF3 #########

    orf3 = []
    orf3_proteins = ''


    for aa3 in range(2,len(seq),3):     #same methodology as above, but again begin iterating one character down.
        orf3.append(seq[aa3:aa3+3])

    for codon3 in orf3:
        if len(codon3) != 3:
            orf3.remove(codon3)

    for i3 in orf3:
        if i3 in aminos.keys():
            orf3_proteins += aminos[i3]

    for i in range(len(orf3_proteins)):
        for j in range(len(orf3_proteins)):
            if orf3_proteins[i] == 'M':
                if orf3_proteins[j] == '*':
                    coding_seq.append(orf3_proteins[i:j:])
                    
    return coding_seq
                
####### REVERSE COMPLEMENT #########

def ReverseComplement(seq):
    
    rna = ''

    for nt in seq:
        if nt == 'A':
            rna += 'T'

        elif nt == 'T':
            rna += 'A'

        elif nt == 'G':
            rna += 'C'

        elif nt == 'C':
            rna += 'G'

    rev_com = (rna[::-1])
    
    return rev_com

def IdentifyCodingRegions(coding_seq):
    
    ######### Coding Region ############
    for i in coding_seq:
        if '*' in i:
            coding_seq.remove(i) 

    coding_region = []
    final_region = []

    for i in coding_seq:
        if i.startswith('M') and '*' not in i:
            coding_region.append(i)

    for i in coding_region:
        if i not in final_region:
            final_region.append(i)

#     for i in final_region:
#         print(i)
        
    return final_region


```
