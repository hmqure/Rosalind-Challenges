```python
def GetSequence(file):
    
    with open(file, 'r') as dna:  #open fasta file
        data = dna.readlines()

    seq = ''

    for row in data[1:]:      #iterate through file after header and strip newline characters
        fasta = (row.strip('\n'))
        seq = seq + fasta     #join to empty list

    dna.close()  #close file

    return seq

def Print4merComposition(seq):
    
    import random

    st = 'A C G T'
    n = 4

    letters = st.split(" ")

    permlist=[]
    perm = ""

    permlist.append(random.choices(letters, k=n))

    runs = 0

    runs = len(letters)**4

    while len(permlist) < runs:
        for j in range(len(permlist)):
            for i in range(n+1):
                y = random.choices(letters, k=4)
                if y not in permlist:
                    permlist.append(y)

    final = []

    for i in permlist:
        final.append(''.join(i))

    for i in final:
        if i == '':
            final.remove(i)

    final = sorted(final, key=lambda word: [letters.index(c) for c in word])

    # for i in final:
    #     print("".join(i))

    kmers = []
    mers = []

    for kmer in range(0,len(seq),1):   #iterate through dna seq and append every three letters to new list as codons
        kmers.append(seq[kmer:kmer+4])

    for i in kmers:
        if len(i) == 4:
            mers.append(i)

    amer = []
    cmer = []
    gmer = []
    tmer = []

    for i in mers:
        if i.startswith('A'):
            amer.append(i)
        if i.startswith('C'):
            cmer.append(i)
        if i.startswith('G'):
            gmer.append(i)
        if i.startswith('T'):
            tmer.append(i)

    amer.sort()
    cmer.sort()
    gmer.sort()
    tmer.sort()

    amer_final = []

    for i in amer:
        if i not in amer_final:
            amer_final.append(i)

    amer_count = []
    cmer_count = []
    gmer_count = []
    tmer_count = []


    for i in final:
        if i.startswith('A'):
            amer_count.append(amer.count(i))

    for i in final:
        if i.startswith('C'):
            cmer_count.append(cmer.count(i))

    for i in final:
        if i.startswith('G'):
            gmer_count.append(gmer.count(i))

    for i in final:
        if i.startswith('T'):
            tmer_count.append(tmer.count(i))   

    joinedlist = amer_count + cmer_count + gmer_count + tmer_count  

    print(*joinedlist, sep=' ')
```

```python
seq = GetSequence('rosalind_kmer.txt')
Print4merComposition(seq)
```
