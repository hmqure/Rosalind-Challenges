```python
def Getn(file):
    
    with open(file, 'r') as dna:  #open fasta file
        data = dna.readlines()
    
    n = int(data[0].strip('\n'))
    
    dna.close()  #close file
    
    return n 

def PrintSignedPermutations(n):

    import random

    #slow but works 

    # n = 5

    posfactor = []
    negfactor = []

    for i in range(1,n+1):
        posfactor.append(i)
        negfactor.append(i * -1)

    factors = posfactor + negfactor

    force = 1

    for i in posfactor:
        force = force * i

    runs = force * (2**n)

    factorial = []

    while len(factorial) < runs:
        for i in range(len(posfactor)):
            y = random.choices(factors, k=n)
            if y not in factorial:
                if len(list(set(map(abs, y)))) == n:
                    factorial.append(y)
    #                 factorial.append(y)
    
    print(len(factorial))
    factorial.sort()
    for i in factorial:
        print(*i, sep=' ') 
```

```python
n = Getn('rosalind_sign.txt')
PrintSignedPermutations(n)
```
